```python
import pandas as pd
df = pd.read_csv("orders.csv")
df.head()
```

```python
# rename columns names. Make them lower case and replace space with underscore
# df.rename(columns={'Order Id':'order_id', 'City':'city'})
# df.columns = df.columns.str.lower() # Change all uppercase letters to lowercase
# df.columns = df.columns.str.replace('_', '')  
df.columns = (
    df.columns
        .str.strip() # Remove spaces before and after string
        .str.lower()  # # Change all uppercase letters to lowercase
        .str.replace(' ', '_', regex=False)  # Change all ‘ ’ to ‘_’
)
df.head(5)
```

```python
# derive new columns discount, sale price and profit
df['discount'] = df['list_price']*df['discount_percent']*.01 # .01 Convert percentage to decimal
df['sale_price'] = df['list_price'] - df['discount']
df['profit'] = df['sale_price'] - df['cost_price']
df
```

```python
# convert order date from object data type to datetime
# df.dtypes # View data types
df['order_date'] = pd.to_datetime(df['order_date'], format = "%Y-%m-%d") # style: 2001-01-01
```

```python
# drop cost price list price and discount percent columns
df.drop(columns = ['list_price', 'cost_price', 'discount_percent'], inplace = True) # inplace means to directly modify the original dataframe
```
